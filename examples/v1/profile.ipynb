```python
from heavyedge import ProfileData
import matplotlib.pyplot as plt

with ProfileData("../../datasets/v1/profiles/dataset1/001.h5") as data:
    Ys, Ls, _ = data[:]
with ProfileData("../../datasets/v1/profiles/dataset1/001-Mean.h5") as data:
    (mean_Y,), (mean_L,), _ = data[:]

for Y, L in zip(Ys, Ls):
    plt.plot(Y[:L], color="gray", alpha=0.5)
plt.plot(mean_Y[:mean_L])
plt.show()
```
